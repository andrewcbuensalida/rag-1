```python
from langchain.vectorstores.cassandra import Cassandra
from langchain.indexes.vectorstore import VectorStoreIndexWrapper
from langchain.llms import OpenAI
from langchain.embeddings import OpenAIEmbeddings

from cassandra.cluster import Cluster
from cassandra.io.libevreactor import LibevConnection

from cassandra.auth import PlainTextAuthProvider

from datasets import load_dataset

# Set up the connection to Cassandra
cloud_config = {
  'secure_connect_bundle': 'secure-connect-an-honest-observer.zip'
}

ASTRA_CLIENT_ID = os.environ.get("ASTRA_CLIENT_ID")
print('*********Example ASTRA_CLIENT_ID: ', ASTRA_CLIENT_ID)
ASTRA_CLIENT_SECRET = os.environ.get("ASTRA_CLIENT_SECRET")


auth_provider = PlainTextAuthProvider(ASTRA_CLIENT_ID, ASTRA_CLIENT_SECRET)
cluster = Cluster(connection_class=LibevConnection,cloud=cloud_config, auth_provider=auth_provider)
session = cluster.connect()

# Connect to the Astra database
session.set_keyspace('keyspace_name')
```
